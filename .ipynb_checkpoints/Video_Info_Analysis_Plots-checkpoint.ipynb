```python
import pandas as pd
video_df = pd.read_csv('Magic_Ingredients_Videos_Stats_V2.csv')
```

# Data Pre-processing

```python
video_df.isnull().any() # check for NULL values
```

```python
video_df.dtypes # check datatype
```

```python
# convert columns with count to numeric
numeric_cols = ['viewCount','likeCount', 'dislikeCount', 'commentCount']
video_df[numeric_cols] = video_df[numeric_cols].apply(pd.to_numeric, axis = 1) # axis=1 means apply to each row
```

```python
video_df.dtypes # check datatype
```

```python
# Publish day in the week
from dateutil import parser
video_df['publishedAt'] = video_df['publishedAt'].apply(lambda x: parser.parse(x)) 
video_df['pushblishDayName'] = video_df['publishedAt'].apply(lambda x: x.strftime("%A")) 
```

```python
video_df.head(2)
```

```python
# convert duration to seconds
import isodate
video_df['durationsecs'] = video_df['duration'].apply(lambda x: isodate.parse_duration(x))
video_df['durationsecs'] = video_df['durationsecs'].astype('timedelta64[s]')
video_df.loc[0:2, ['durationsecs', 'duration']]
```

```python
import os
from os import path
# get data directory (using getcwd() is needed to support running example in generated IPython notebook)
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()

#stopwords_path = d + '/wc_cn/stopwords_cn_en.txt'
# Chinese fonts must be set
font_path = d + '/SourceHanSerifK-Light.otf'

stop_words = set(stopwords.words('english'))
video_df['title_no_stopwords'] = video_df['title'].apply(lambda x: [item for item in str(x).split() if item not in stop_words])

all_words = list([a for b in video_df['title_no_stopwords'].tolist() for a in b])
all_words_str = ' '.join(all_words) 

def plot_cloud(wordcloud):
    plt.figure(figsize=(30, 20))
    plt.imshow(wordcloud) 
    plt.axis("off");

wordcloud = WordCloud(font_path=font_path, width = 2000, height = 1000, random_state=1, background_color='black', 
                      colormap='viridis', collocations=False).generate(all_words_str)
plot_cloud(wordcloud)
```
